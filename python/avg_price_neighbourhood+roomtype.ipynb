```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el CSV exportado
df = pd.read_csv('Avg_price_neighbourhood+roomtype.csv')

# Crear figura con 2 subplots (1 fila, 2 columnas)
fig, axes = plt.subplots(1, 2, figsize=(16,6))  # ancho x alto

# -------- Gráfico 1: solo por neighbourhood_group ----------
sns.barplot(data=df, x='neighbourhood_group', y='avg_price', palette='viridis', ax=axes[1])
axes[1].set_ylabel('Average Price ($)')
axes[1].set_xlabel('Neighbourhood Group')

for p in axes[1].patches:
    height = p.get_height()
    axes[1].annotate(f"${height:.2f}", 
                     (p.get_x() + p.get_width()/2., height + 5),
                     ha='center', va='bottom')

# -------- Gráfico 2: agrupado por room_type ----------
sns.barplot(data=df, x='neighbourhood_group', y='avg_price', hue='room_type', ax=axes[0])
axes[0].set_ylabel('Average Price ($)')
axes[0].set_xlabel('Neighbourhood Group')
axes[0].legend(title='Room Type')

plt.tight_layout()  # Ajusta todo para que no se solape
plt.show()
```
